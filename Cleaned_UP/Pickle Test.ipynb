```python
import random

# Assuming you have a dictionary called my_dict
# Replace my_dict with your actual dictionary

X_train_data = []  # List to store 80% of the elements
X_test_data = [] # List to store 20% of the elements

y_train_data = []
y_test_data = []

master_keys = list(master_team_dict.keys())  # Create a list of keys
random.shuffle(master_keys)  # Shuffle the list in-place
for key in master_keys:
    train_size = generate_random_with_probability(0.8)
    # Shuffle the elements
    
    if train_size:
        # Append 80% of elements to X_train_data
        X_train_data.append(torch.tensor(master_team_dict[key]).float())
        y_train_data.append(torch.tensor(y_dict[key]).float())
        
    else:
    # Append 20% of elements to X_test_data
        X_test_data.append(torch.tensor(master_team_dict[key]).float())
        y_test_data.append(torch.tensor(y_dict[key]).float())

# Print the lengths of the resulting sets
print("Length of X_train_data:", len(X_train_data))
print("Length of X_test_data:", len(X_test_data))
print("\n")
print("Length of y_train_data:", len(y_train_data))
print("Length of y_test_data:", len(y_test_data))
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
input_size=len(master_team_dict["BUF_11.0_2021"])
output_size=len(y_dict["BUF_11.0_2021"])

class YourModel(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(YourModel, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        out = self.fc1(x)
        out = self.relu(out)
        out = self.fc2(out)
        out = self.relu(out)  # Applying ReLU to the output layer
        return out

# Instantiate the model
model = YourModel(input_size=input_size, hidden_size=123, output_size=output_size)

# Example usage of L2 regularization term in the loss
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Forward pass
inputs = torch.randn(32, input_size)  # Example input data, batch size 32
outputs = model(inputs)
labels = torch.randint(0, output_size, (32,))  # Random labels for the example

# Calculate the loss
loss = criterion(outputs, labels)

# Backward and optimize
optimizer.zero_grad()
loss.backward()
optimizer.step()

# Print the model to see the architecture
print(model)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_squared_error

# List of positions
positions = ["QB_1", "RB_1", "LWR_1", "RWR_1", "SWR_1", "TE_1"]

# Assuming you have X_train_data, X_test_data, y_train_data, y_test_data, model, optimizer, and criterion

# Lists to store training and validation losses
training_losses = []
validation_losses = {pos: [] for pos in positions}  # Dictionary to store MSE for each position

# Training and Validation
num_epochs = 10  # Set the number of epochs

# Function to preprocess the data (handling NaN and infinity)
def preprocess_data(data):
    # Replace NaN and infinity with zeros
    data = np.nan_to_num(data)
    return data

for epoch in range(num_epochs):
    model.train()  # Set the model to training mode

    total_loss = 0.0

    # Inside the training loop
    for i, data in enumerate(X_train_data):
        # Assuming X_train_data is a list of training samples
        inputs = torch.tensor(preprocess_data(data)).float()
        # Assuming y_train_data is a list of corresponding labels
        labels = torch.tensor(preprocess_data(y_train_data[i])).float()

        optimizer.zero_grad()
        outputs = model(inputs)

        # Ensure outputs and labels have compatible dimensions
        assert outputs.shape[0] == labels.shape[0], "Output and label dimensions don't match"

        loss = criterion(outputs, labels)
        loss.backward()

        # Apply gradient clipping
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1)

        optimizer.step()

        total_loss += loss.item()

        # Print loss and completion percentage
        print(f"Epoch [{epoch+1}/{num_epochs}], Progress: {i/len(X_train_data)*100:.2f}%, Loss: {loss.item():.4f}", end="\r")

    average_loss = total_loss / len(X_train_data)
    training_losses.append(average_loss)
    print(f"\nEpoch [{epoch+1}/{num_epochs}], Average Training Loss: {average_loss:.4f}")

    # Validation
    model.eval()  # Set the model to evaluation mode

    with torch.no_grad():
        for i, data in enumerate(X_test_data):
            inputs = torch.tensor(preprocess_data(data)).float()
            labels = torch.tensor(preprocess_data(y_test_data[i])).float()

            outputs = model(inputs)

            # Ensure there are no NaN or infinity values
            assert torch.isfinite(outputs).all(), "Model output has NaN or infinity values."

            # Ensure labels and outputs have the same number of samples
            assert labels.shape[0] == outputs.shape[0], "Number of samples mismatch"

            # Calculate MSE per output element and store in the dictionary
            mse_per_output = mean_squared_error(labels, outputs, multioutput='raw_values')

            # Handle single-value MSE
            if mse_per_output.size == 1:
                validation_losses[positions[0]].append(mse_per_output.item())
            else:
                for j, pos in enumerate(positions):
                    validation_losses[pos].append(mse_per_output[j])
```

```python
import torch

def predict(model, X_data):
    """
    Predict using a pre-trained PyTorch model.
    
    Parameters:
    model (torch.nn.Module): Trained PyTorch model.
    X_data (torch.Tensor): Input data for prediction.
    
    Returns:
    torch.Tensor: Model predictions.
    """
    # Set the model to evaluation mode
    model.eval()
    
    # Convert input data to torch tensor and float
    inputs = torch.tensor(X_data).float()
    
    # Make predictions
    with torch.no_grad():
        outputs = model(inputs)
    
    return outputs.tolist()

# Example usage
# Assuming you have a trained model called 'trained_model' and input data 'X_test'
# predictions = predict(trained_model, X_test)
```
